```python
import HalpinTsai as HT
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
npontos=20

Vf = np.linspace(0,1,npontos)

Ef=233
Em=4.62
nuf=0.2
num=0.36
Gm = Em/(2*(1 + num))
Gf= Ef/(2*(1+nuf))
EL=HT.calc_EL(Vf,Ef,Em)
```

```python
d=1
l=10
EL2=HT.calc_EL_fibras_curtas(Vf,Ef,Em,l,d)
```

```python
plt.figure()
p1 = plt.plot(Vf,EL,linestyle="-",marker="x",label="$E_L$ fibras contínuas")
p1 = plt.plot(Vf,EL2,linestyle="--",marker="o",label="$E_L$ fibras curtas alinhadas")
plt.axis("tight") # Fit the axis tightly to the plot
plt.grid("on")
plt.legend(loc="upper center",fancybox="true") # Create a legend of all the existing plots using their labels as names
plt.xlabel("$V_f$")
plt.ylabel("$E_T$")
plt.title("Módulo de elasticidade longitudinal")
plt.show()
```

```python
ET=HT.calc_ET(Vf,Ef,Em,2)
GLT=HT.calc_GLT(Vf,Ef,Em,2)
E=2/8*EL2+5/8*ET
G=1/8*EL2+1/4*ET
```

```python
plt.figure()
p1 = plt.plot(Vf,EL2,linestyle="-",marker="x",label="$E_L$ fibras curtas alinhadas")
p1 = plt.plot(Vf,E,linestyle="--",marker="o",label="$E$ fibras curtas aleatórias")
plt.axis("tight") # Fit the axis tightly to the plot
plt.grid("on")
plt.legend(loc="upper center",fancybox="true") # Create a legend of all the existing plots using their labels as names
plt.xlabel("$V_f$")
plt.ylabel("$E_T$")
plt.title("Módulo de elasticidade")
plt.show()
```

```python
plt.figure()
p1 = plt.plot(Vf,GLT,linestyle="-",marker="x",label="$G_{LT}$ fibras curtas alinhadas")
p1 = plt.plot(Vf,G,linestyle="--",marker="o",label="$G$ fibras curtas aleatórias")
plt.axis("tight") # Fit the axis tightly to the plot
plt.grid("on")
plt.legend(loc="upper center",fancybox="true") # Create a legend of all the existing plots using their labels as names
plt.xlabel("$V_f$")
plt.ylabel("$G_{LT}$")
plt.title("Módulo de elasticidade")
plt.show()
```
