```python
from IPython.display import display, Markdown
import numpy as np
import matplotlib.pyplot as plt
```

Um compósito é fabricado com fibras de vidro (diâmetro = 0,03 mm) em uma matriz de resina epóxi. Todas as fibras estão alinhadas paralelamente à direção da aplicação da carga. A fração volumétrica de fibras é de 40%. Suponha que a matriz se comporte como um material plástico rígido com um limite de escoamento de 28 MPa e que $E_f$ = 70 GPa e $E_m$ = 3,5 GPa.

 Determine o comprimento da zona de transferência de carga $l_t$, para tensões compostas $\sigma_c$ = 70 MPa e $\sigma_c$ = 210 MPa.

Traçar a tensão normal na fibra e a tensão de cisalhamento interfacial ao longo do comprimento da fibra para fibras com comprimento $l = 1/2l_t$, $l_t$ e 4$l_t$. Calcular também a tensão média na fibra em cada um dos casos.



```python
# Unidades: dimensões em mm, forças em Newton
#           tensões e módulos de elasticidade em MPa
Ef=70e3
Em=3.5e3
Vf=0.4
Vm=1-Vf
d=0.03
r=d/2
Ec=Ef*Vf+Em*Vm
sigmay=28
tauy=sigmay/2
```

```python
sigmac=70
lt=(Ef/Ec)*sigmac/(2*tauy)*d
display(Markdown(rf"""$\sigma_c$ = {sigmac} MPa

$l_t$ = {lt:5.2} mm"""))
```

```python
sigmac=210
lt=(Ef/Ec)*sigmac/(2*tauy)*d
display(Markdown(rf"""$\sigma_c$ = {sigmac} MPa

$l_t$ = {lt:5.2} mm"""))
```

```python
npontos=20
l=lt/2
z1=np.linspace(0,l/2,npontos)
z2=np.linspace(l/2,l,npontos)
sigmaf1=2*tauy*z1/r
sigmaf2=2*tauy/r*(l-z2)
plt.figure()
plt.plot(z1,sigmaf1,"ro-")
plt.plot(z2,sigmaf2,"ro-")
plt.show()
sigmafmax=sigmac*Ef/Ec
display(Markdown(rf"""$(\sigma_f)_m$ = {sigmafmax:8.4} MPa"""))
```

```python
npontos=20
l=lt
z1=np.linspace(0,l/2,npontos)
z2=np.linspace(l/2,l,npontos)
sigmaf1=2*tauy*z1/r
sigmaf2=2*tauy/r*(l-z2)
plt.figure()
plt.plot(z1,sigmaf1,"ro-")
plt.plot(z2,sigmaf2,"ro-")
plt.show()
sigmafmax=sigmac*Ef/Ec
display(Markdown(rf"""$(\sigma_f)_m$ = {sigmafmax:8.4} MPa"""))
```

```python
npontos=20
l=4*lt
sigmafmax=sigmac*Ef/Ec
z1=np.linspace(0,lt/2,npontos)
z2=np.linspace(lt/2,(4-1/2)*lt,npontos)
z3=np.linspace((4-1/2)*lt,4*lt,npontos)
sigmaf1=2*tauy*z1/r
sigmaf2=sigmafmax*np.ones(npontos)
sigmaf3=2*tauy/r*(l-z3)
plt.figure()
plt.plot(z1,sigmaf1,"ro-")
plt.plot(z2,sigmaf2,"ro-")
plt.plot(z3,sigmaf3,"ro-")
plt.show()
display(Markdown(rf"""$(\sigma_f)_m$ = {sigmafmax:8.4} MPa"""))
```
